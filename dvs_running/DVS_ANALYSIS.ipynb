```python
PLOT_HIST_SIZE = 30
PLOT_KIND = 'box'
SCATTER_PLOT_LIMIT = 50
BUBBLE_SIZE = 300.0
RANK = 30
```

### Define functions to execute with graphs

```python
import seaborn as sns
import numpy as np

def plotHist():
    '''
    Take PLOT_HIST_SIZE's value and genrate histogram from that number
    of random variables
    
    '''
    x = np.random.normal(size = PLOT_HIST_SIZE)
    return sns.distplot(x)
```

```python
plotHist()
```

```python
def bar_box(PLOT_KIND):
    tips = sns.load_dataset("tips")
    figure = sns.catplot(x="day", y="total_bill", kind=PLOT_KIND, data=tips)
    figure.fig
```

```python
bar_box(PLOT_KIND)
```

```python
import matplotlib.pyplot as plt
from numpy.random import rand

def drawScatter(SCATTER_PLOT_LIMIT, BUBBLE_SIZE):
    fig, ax = plt.subplots()
    for color in ['red', 'green', 'blue']:
        x, y = rand(2, SCATTER_PLOT_LIMIT)

        scale = BUBBLE_SIZE * rand(SCATTER_PLOT_LIMIT)

        ax.scatter(x, y, c=color, s=scale, label=color,
                   alpha=0.3, edgecolors='none')

    ax.legend()
    ax.grid(True)

    plt.show()
```

```python
drawScatter(SCATTER_PLOT_LIMIT, BUBBLE_SIZE)
```

```python
def howManyHigherThan():
    '''
    Check how many values are greater then provided values in list
    '''
    count = 0
    ourList = [1, 12, 24,3,45 ,54 ,14, 32]
    
    for item in ourList:
        if(RANK < item):
            count +=1
    return count       
```
